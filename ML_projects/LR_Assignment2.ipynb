```python
import tensorflow as tf
from d2l import tensorflow as d2l
```

```python
class MLP(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens, lr, sigma=0.01):
        super().__init__()
        self.save_hyperparameters()
        self.net = tf.keras.models.Sequential([
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(num_hiddens, activation='relu'),
            tf.keras.layers.Dense(num_outputs)])
        
def train_and_plot(learning_rates):
    for l_r in learning_rates:
        model = MLP(num_outputs=10, num_hiddens=256, lr=l_r)
        model.board.yscale='log'
        data = d2l.FashionMNIST(batch_size=256)
        trainer = d2l.Trainer(max_epochs=20)
        trainer.fit(model, data) 
        
        model.summary()

learning_rates = [0.001, 0.01, 0.1, 1, 100, 1000]

train_and_plot(learning_rates)
```
